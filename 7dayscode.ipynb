# # Desafio 7 days of Code da Alura

```python
import pandas as pd

url_dados = 'https://github.com/fabiollg/7dayscode/blob/main/archive.zip?raw=true' # obtém os dados do repositório no Github, baixados do Kaggle.

dados_raw = pd.read_csv(url_dados, compression = 'zip') # lê os dados csv.

pd.set_option('display.max_columns',None) # possibilita exibir todas as colunas do dataset.

dados_raw
```

# Filtrando os dados

```python
dados_filtrado = dados_raw.drop(columns=['Unnamed: 0','track_id','explicit']) #eliminando colunas que não agregam valor.
                             
dados_filtrado
```
